```python
import pandas as pd
import matplotlib.pylab as plt

filename = 'date_result.csv'
outfile = 'out_plot_new.csv'

n_points = 9


df = pd.read_csv(filename)

df['sq_x_y'] = (df.x**2 + df.y**2)**0.5
df_1 = df.shift(1)  #数据向后移动一行，

data_test = []
for i in range(0, len(df_1), 9):
    df_temp_1 = df_1.sq_x_y[i:i+9]
#     print(df_temp_1)
    df_temp_2 = df_1.yaw[i:i+9]
#     print(df_temp_2)
    entry = list(df_temp_1) + list(df_temp_2)
    data_test.append(entry)
    
columns_name = ['p_'+str(i+1) for i in range(n_points)]+['a_'+str(i+1) for i in range(n_points)]

df_out = pd.DataFrame(data_test, columns = columns_name)

df_out.to_csv(outfile, index = False)

```
